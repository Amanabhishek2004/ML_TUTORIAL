```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train.csv")
df
```

```python
a = df["Survived"].groupby(df["Sex"]).count()
labels = "Survived", "Died"
sizes = [a[1], a[0]]  # Adjust the order based on your DataFrame structure

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Survival Distribution based on sex')
plt.show()
```

```python
a = [1,45,3,6,"A"]
b = []
count = 0
for i in range(1000):
   
   
    for j in a:
       if count == 891:
           break
       else:
            b.append(j)
            count+=1


df["numbers"] = b
df
```

```python
plot = df["numbers"].value_counts().plot(kind = "bar")
```

```python
# transforming the data to numeric form

df["numbers_numeric"] = pd.to_numeric(df["numbers"] , errors="coerce" , downcast="integer") 
df
```

```python
df["numbers_charcters"] = np.where(df["numbers_numeric"].isnull() , df["numbers"] , np.nan)
df
```

```python
df["cabin_num"] = df["Cabin"].str.extract(str("(/d+)"))
df["cabin_cat"] = df["Cabin"].str[0]
df
```

```python
df["ticket_num"] = df["Ticket"].apply(lambda s:s.split()[-1])
df["ticket_num"] = pd.to_numeric(df["ticket_num"] ,errors="coerce" , downcast="integer")
df
```

<h4>Extract Year</h4>

```python
date["date_year"] = date["date"].dt.year
date.head()
```

```python
date["date_month_name"] = date["date"].dt.month_name()
date["date_month_no"] = date["date"].dt.month
date["date_day"] = date["date"].dt.day
date["date_dow"] = date["date"].dt.day_of_week
date["date_dow_day_name"] = date["date"].dt.day_name()
date.head()
```

```python
date["is_weekend"] = np.where(date["date_dow_day_name"].isin(["Sunday" , "Saturday"]),1,0)
date
```
