```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# bivariate analysis ->  if you just have two variables for analysis
price = [123500,200000,120000,3451267]
year = [2012,2016,2017,2018]
```

```python
plt.plot(year,price)
# got a graph of price vs year
```

```python
batsman = pd.read_csv("sharma-kohli.csv")
batsman
```

```python
plt.plot(batsman["index"],batsman["V Kohli"],color = "green" ,linestyle = "dashed" , linewidth = "5" , marker = "D" , markersize = "10", label = "Virat")
plt.plot(batsman["index"],batsman["RG Sharma"], label = "Rohit")
#  even the html color codes be pasted here
plt.title("Rohit shrama VS Virat Kohli")
plt.xlabel("Seasons")
plt.ylabel("Runs scored")
plt.legend(loc = "upper right")
plt.show()
```

```python
# setting the limits
price = [123500,200000,120000,3451267000000000]
year = [2012,2016,2017,2018]
plt.plot(year,price)
```

```python
# for avoiding the straight line thing
price = [123500,200000,120000,3451267000000000]
year = [2012,2016,2017,2018]
plt.plot(year,price)

plt.ylim(0,1000000)
```

```python
plt.plot(batsman["index"],batsman["V Kohli"],color = "green" ,linestyle = "dashed" , linewidth = "5" , marker = "D" , markersize = "10", label = "Virat")
plt.plot(batsman["index"],batsman["RG Sharma"], label = "Rohit")
#  even the html color codes be pasted here
plt.title("Rohit shrama VS Virat Kohli")
plt.xlabel("Seasons")
plt.ylabel("Runs scored")
plt.legend(loc = "upper right")
plt.grid()
plt.show()
```

```python
tips = sns.load_dataset("tips")

tips
```

```python
# this is a slower technique

plt.scatter(tips["total_bill"],tips["tip"],s=tips["size"]*20)
```

```python
# this is quite fast
#  can be used at the time of a big data set to train am model
plt.plot(tips["total_bill"],tips["tip"],"o")
```

** Bar graphs

```python
children = [10,20,30,50]
color = ["red", "blue","green","yellow"]
```

```python
plt.bar(color,children , color = "green")
```

```python
df = pd.read_csv("batsman_season_record.csv")
```

```python
plt.bar(df["batsman"],df["2015"], width = 0.1)
```

```python

plt.bar(np.arange(df.shape[0])-0.2, df["2015"], width=0.2, color="red", label="2015")
plt.bar(np.arange(df.shape[0]), df["2017"], width=0.2, color="yellow", label="2017")
plt.bar(np.arange(df.shape[0])+0.2, df["2016"], width=0.2, color="blue", label="2016")

plt.show()
```
