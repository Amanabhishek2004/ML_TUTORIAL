```python
import numpy as np
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

```python
x , y= make_regression(n_features=1 , n_samples=4 , n_targets=1 , n_informative=1 , noise =80 , random_state=13 )
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#  OLS
reg = LinearRegression()
reg.fit(x , y)
```

```python
m = reg.coef_
```

```python
b = reg.intercept_
```

```python
plt.plot(x , reg.predict(x) , color = "red")
plt.scatter(x,y)
```

```python
y_pred = ((78.35*x)+0).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x , reg.predict(x) , color = "red")
plt.plot(x , y_pred , color = "yellow")
```

```python
# GD
b=0
loss_slope = -2*np.sum(y-m*x.ravel()-b)
loss_slope
```

```python
# bnew = bold - lr*loss_slope
lr = 0.1
b = b - loss_slope*lr
b 
```

```python
#  SOLVING THE PROBLEM ITERATIVELY
b=0
lr = 0.01
color = ["red" , "blue" , "purple" , "yellow" , "green" , "red" , "indigo" ]
for i in range(7):
 
  loss_slope = -2*np.sum(y-m*x.ravel()-b)
  b = b - lr*loss_slope 
  y_pred = (m*x+b).reshape(4)
  plt.scatter(x,y)
  plt.plot( x , y_pred , color = color[i] )
```
