```python

# PLAY STORE

import pandas as pd
from bs4 import BeautifulSoup

def data_frame_creator(page):
    soup = BeautifulSoup(page, "lxml")
    data = (soup.find_all("div", class_="cXFu1"))
    rating = []
    name = []

    for i in data:
        rating_value = i.find("span", class_="w2kbF")
        name_value = i.find("span", class_="DdYX5")


        if (rating_value)  and (name_value)  :
            rating.append(rating_value.text)
            name.append(name_value.text)
        
        if (name_value)!=None and(rating_value) == None:
            rating.append(0)
            name.append(name_value.text)


    # Ensure both arrays are the same length


    data = {
        "app_name": name,
        "rating": rating,
    }

    df = pd.DataFrame(data)
    return df

data_frame_creator(page1)
```

```python
def add_app_id(page):
    soup = BeautifulSoup(page, "lxml")
    data = soup.find_all("a", class_="Si6A0c Gy4nib")
    links = []

    for i in data:
        link = i.get("href")
        i = link.find("id")
        if i != -1:  # Check if 'id' exists in link
            app_id = link[i + 3:]
            links.append(app_id)

    return links


def get_images(data):
   images = []
   for i in data:
      page =  requests.get(f"https://play.google.com/store/apps/details?id={i}&hl=en-IN").text

      soup = BeautifulSoup(page, "lxml")
      value = soup.find_all("div" , class_ = "Mqg6jb Mhrnjf")
      if len(value)!=0:
       image_tag = value[0].find("img")
       images.append(image_tag.get("src"))
      else:
       images.append(0)
   return images
```

```python
def final_output(array):
    df = None

    for query in array:
        page = requests.get(f"https://play.google.com/store/search?q={query}&c=apps", headers=headers).text
        if df is None:
            df = data_frame_creator(page)
            df["app_id"] = add_app_id(page)

        else:
            new_df = data_frame_creator(page)
            new_df["app_id"] = add_app_id(page)[:len(df["app_name"])]
            df = pd.concat([df, new_df], ignore_index=True)

    # Drop duplicates and update the DataFrame
            
    df = df[df['app_name'].apply(lambda x: 'JEE' in x.split())][df["rating"].astype(float)>4.3]
    df = df.drop_duplicates(keep="first").reset_index(drop=True)
    df["images"] = get_images(df["app_id"].values)
    return df                     
```

```python
df = final_output(array)
df
```
